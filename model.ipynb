```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
gestures = pd.read_csv("gestures.csv")
gestures
```

```python
ges = ["up", "down", "right", "left", "stop", "back", "forward"]
print(sorted(ges))
for i in ges:
    g = gestures[gestures["gesture"] == i]
    right = g[g["handedness"] == "Right"]
    left = g[g["handedness"] == "Left"]
    print(f"Gesture: {i}\nTotal: {g.shape[0]}\tRight: {right.shape[0]}\tLeft: {left.shape[0]}")
```

```python
gestures.isna().sum()
```

```python
gestures["gesture"] = gestures["gesture"].map({'back': 0, 'down': 1, 'forward': 2, 'left': 3, 'right': 4, 'stop': 5, 'up': 6})
gestures["handedness"] = gestures["handedness"].map({"Left": 0, "Right": 1})
gestures
```

```python
X = gestures.drop("gesture", axis=1)
X
```

```python
y = gestures["gesture"]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=69420, shuffle=True, stratify=y)
```

```python
model = Sequential([
    Dense(32, activation="relu"),
    Dense(64, activation="relu"),
    Dense(32, activation="relu"),
    Dense(7)
])
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
model.fit(X_train, y_train, epochs=100)
```

```python
test_loss, test_acc = model.evaluate(X_test,  y_test, verbose=2)
print('Test loss:', test_loss)
print('Test acc:', test_acc)
```

```python
model.fit(X, y, epochs = 100)
model.summary()
```

```python
probability_model = tf.keras.Sequential([model, tf.keras.layers.Softmax()])
```

```python
probability_model.save("asa.keras")
```
