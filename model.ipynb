```python
import pandas as pd
```

```python
gestures = pd.read_csv("gestures.csv")
gestures
```

```python
ges = ["up", "down", "right", "left", "stop", "flip"]
for i in ges:
    g = gestures[gestures["gesture"] == i]
    right = g[g["handedness"] == "Right"]
    left = g[g["handedness"] == "Left"]
    print(f"Gesture: {i}\nTotal: {g.shape[0]}\tRight: {right.shape[0]}\tLeft: {left.shape[0]}")
```
